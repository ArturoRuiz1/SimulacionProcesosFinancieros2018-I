$$f(x)=\begin{cases}\frac{2}{(c-a)(b-a)}(x-a), & a\leq x \leq b\\ \frac{-2}{(c-a)(c-b)}(x-c),& b\leq x \leq c \end{cases}$$

```python
def f_trian(x,a,b,c):
    if x>=a and x<b:
        f=(2*(x-a))/((c-a)*(b-a))
    elif x>=b and x<=c:
        f=(-2*(x-c))/((c-a)*(c-b))
    else:
        f=0
    return f
# %matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
a=1; b=3; c=5
M = 2/(c-a)
f=[]
x = np.arange(0.5,6,0.1)
for i in range(len(x)):
    f.append(f_trian(x[i],a,b,c))
plt.plot(x,f)
N = 300
i=0
while i<=N:
    u1 = np.random.uniform(0,M,1)
    u2 = np.random.uniform(a,c,1)
    if u1<=f_trian(u2,a,b,c):
        plt.plot(u2,u1,'bo')
        i+=1
    else:
        plt.plot(u2,u1,'o')
plt.show()
```
