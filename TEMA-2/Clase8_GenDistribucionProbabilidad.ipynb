```python
import numpy as np
import matplotlib.pyplot as plt
b = 3
N = 10**6
x = np.random.exponential(b,N)
x2 = np.log(1-np.random.random(N))/(-1/b)
plt.hist(x,100,normed=True)
plt.xlabel('valores aleatorios')
plt.ylabel('probabilidad')
plt.title('histograma')
plt.show()
print(np.mean(x))
```
